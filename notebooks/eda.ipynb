```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

filepath = Path("../data/processed/flight_log.csv")
flights= pd.read_csv(filepath)
```

```python
cm = flights.select_dtypes(include=np.number).corr(min_periods=15)
plt.figure(figsize=(30,20))
sns.heatmap(cm, annot=True, cmap='coolwarm',fmt='.2f')
plt.title("Correlation heatmap")
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(data=flights, x='Delay Carrier (Minutes)', y='Departure delay (Minutes)')
plt.title("Carrier Delay vs Departure Delay")
plt.show()

plt.figure(figsize=(20,15))
sns.scatterplot(data=flights, x='Delay Weather (Minutes)', y='Departure delay (Minutes)')
plt.title("Weather Delay vs Departure Delay")
plt.show()

plt.figure(figsize=(20,15))
sns.scatterplot(data=flights, x='Delay Late Aircraft Arrival (Minutes)', y='Departure delay (Minutes)')
plt.title("Late Aircraft Delay vs Departure Delay")
plt.show()
```
